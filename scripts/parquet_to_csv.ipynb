```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("IPL-Batch-Analytics") \
    .getOrCreate()
```

```python
paths = {
    "player_wise_runs": "hdfs:///ipl/curated/player_wise_runs",
    "player_wise_fours": "hdfs:///ipl/curated/player_wise_fours",
    "player_wise_sixes": "hdfs:///ipl/curated/player_wise_sixes",
    "player_wise_wickets": "hdfs:///ipl/curated/player_wise_wickets",
    "player_wise_catches": "hdfs:///ipl/curated/player_wise_catches",
    "player_wise_runouts": "hdfs:///ipl/curated/player_wise_runouts",
    "player_wise_stumpings": "hdfs:///ipl/curated/player_wise_stumpings",
    "player_matches": "hdfs:///ipl/curated/player_matches"
}

for name, hdfs_path in paths.items():
    df = spark.read.parquet(hdfs_path)
    df.coalesce(1).write \
        .mode("overwrite") \
        .option("header", "true") \
        .csv(f"/home/talentum/ipl_bigdata_project/ipl_exports_csv/{name}")

print("ALL player-wise Parquet datasets converted to CSV successfully")
```
